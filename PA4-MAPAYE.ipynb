```python
import pandas as pd
```

```python
# Read the Excel file "board2.xlsx" and load it into a DataFrame
df=pd.read_excel("board2.xlsx")

# Output of DataFrame
df
```

### 1. Create the following data frames based on the format provided:

#### a.) Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# Filters the rows based on the conditions and selects the specific columns
Instru = df.loc[(df['Track'] == 'Instrumentation')
                &(df['Hometown'] == 'Luzon')
                &(df['Electronics']>70)] [['Name','GEAS','Electronics']]

# Output of Instru
Instru
```

#### b.) Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
# Calculates the Averages of Math, Electronics, GEAS, and Communication
# Also adds an additional Average column to display Average
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

# Output
df
```

```python
# Locates 'Name', 'GEAS', 'Average' for 'Mindanao' females with Average >= 55
Mindy = df.loc[(df['Hometown']=='Mindanao')
               &(df['Gender']=='Female')
               &(df['Average']>=55)] [['Name','GEAS','Average']]

# Output
Mindy
```

### 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
import matplotlib.pyplot as plt
```

```python
# Set the figure size for the plot
plt.figure(figsize=(15, 6))

# Creates a bar plot for the Averages of the Gender, Track, and Hometown
plt.bar(df['Gender'] , df['Average'])
plt.bar(df['Track'], df['Average'])
plt.bar(df['Hometown'], df['Average'])
```
